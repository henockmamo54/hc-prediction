```python
import pandas as pd
import numpy as np
import warnings
from matplotlib import pyplot as plt
from sklearn import metrics as m
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE  # doctest: +NORMALIZE_WHITESPACE
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from collections import Counter

warnings.filterwarnings("ignore")
randomseed = 7
```

```python
#diabetes required columns 
req_cols=["Unnamed: 0","FIELD_1",'L100800', 'L104600', 'L103000', 'S000300', 'L101700', 'L100700', 'FIELD_33',
            'FIELD_38', 'FIELD_40', 'FIELD_31', 'SEX', 'AGE',"CLASS"]


data1 = pd.read_csv("ThisYearData_actualData_train.txt")[req_cols]

data2 = pd.read_csv("NextYearData_actualData_train.txt")[req_cols]

data2_p = pd.read_csv("Predicted_NextYearData_train.txt")[['Unnamed: 0', "P_FIELD_1",'P_L100800', 'P_L104600', 'P_L103000', 
                                                           'P_S000300', 'P_L101700', 'P_L100700', 'P_FIELD_33','P_FIELD_38',
                                                           'P_FIELD_40', 'P_FIELD_31', 'P_SEX', 'P_AGE','A_CLASS']]

data2_p.columns = data2.columns





data3 = pd.read_csv("ThisYearData_actualData_test.txt")[req_cols]
data4 = pd.read_csv("NextYearData_actualData_test.txt")[req_cols]

data5 = pd.read_csv("Predicted_NextYearData.txt")[['Unnamed: 0', "P_FIELD_1",'P_L100800', 'P_L104600', 'P_L103000', 
                                                           'P_S000300', 'P_L101700', 'P_L100700', 'P_FIELD_33','P_FIELD_38',
                                                           'P_FIELD_40', 'P_FIELD_31', 'P_SEX', 'P_AGE','A_CLASS']]
data5.columns = data4.columns

# data = pd.concat([data1, data2, data3])
# data=data1

data1=data1.drop(columns=["CLASS"])
data3=data3.drop(columns=["CLASS"])

# print(data.shape)
# data.head()
```

```python
temptrain=pd.merge(data1,data2_p,left_on=["FIELD_1","Unnamed: 0"], right_on=["FIELD_1","Unnamed: 0"]
                  ).drop(columns=["Unnamed: 0","FIELD_1"])
print(temptrain.shape)
print(temptrain.columns)
```

```python
temptest=pd.merge(data3,data5,left_on=["FIELD_1","Unnamed: 0"], right_on=["FIELD_1","Unnamed: 0"]
                  ).drop(columns=["Unnamed: 0","FIELD_1"])
```

```python
import random
r = random.Random()
r.seed(42)

extremeDiabetesdata = pd.concat([temptrain[temptrain.CLASS==2],temptrain[temptrain.CLASS==2]])
fpgnewadditionalvalues = []
hbalcnewadditionalvalues = []
for i in range(extremeDiabetesdata.shape[0]):
    fpgnewadditionalvalues.append(r.randint(30, 200))
    hbalcnewadditionalvalues.append(r.randint(10, 150) / 10)

    extremeDiabetesdata.iloc[i, 0] = (
        extremeDiabetesdata.iloc[i, 12] + fpgnewadditionalvalues[i]
    )

    extremeDiabetesdata.iloc[i, 1] = (
        extremeDiabetesdata.iloc[i, 13] + hbalcnewadditionalvalues[i]
    )
    
```

```python
temptrain=pd.concat([extremeDiabetesdata,temptrain])
```

```python
xtrain = temptrain.iloc[:, :-1]
ytrain = temptrain.iloc[:, -1]
xtest = temptest.iloc[:, :-1]
ytest = temptest.iloc[:, -1]
```
